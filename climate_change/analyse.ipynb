# Marketing Analyse

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

marketing_campaign = pd.read_csv("marketing_campaign.csv", sep ='\t')

print(marketing_campaign)
```

```python
marketing_campaign.columns = (marketing_campaign.columns
                .str.replace('(?<=[a-z])(?=[A-Z])', '_', regex=True)
                .str.lower()
             )
print(marketing_campaign)
```

```python
marketing_campaign.dropna(axis = 0)
```

```python
marketing_campaign["children_count"] = (marketing_campaign["kidhome"] + marketing_campaign["teenhome"])

marketing_campaign["num_total_purchases"] = (marketing_campaign["num_deals_purchases"] + marketing_campaign["num_web_purchases"] + marketing_campaign["num_catalog_purchases"])
```

```python
plt.hist(marketing_campaign["income"], bins=20)
plt.show()
```

```python
plt.scatter(marketing_campaign["year_birth"], marketing_campaign["income"])
plt.show()
```

```python
filtered_df = marketing_campaign[(marketing_campaign['income'] < 100000) & (marketing_campaign['year_birth'] > 1920)]
plt.scatter(filtered_df['year_birth'],filtered_df['income'])
plt.ylabel('Income')
plt.xlabel('Year of Birth')
plt.title('Scatter Plot of Income vs. Year of Birth')
plt.show()
```

```python
Edu_Income = marketing_campaign.groupby(['education'])['income'].mean()
Edu_Income.plot(kind="barh", title="Average Income from Educational Level", xlabel="Income in USD", ylabel="Educational Level")
```
